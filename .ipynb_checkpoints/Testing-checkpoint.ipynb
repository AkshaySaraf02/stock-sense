```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

import pickle
```

```python
headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246"}
```

```python
def latest_5_news(company_name):
    url = f"https://economictimes.indiatimes.com/topic/{company_name}"
    response = requests.get(url=url, headers=headers)

    output = []
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        for i, news in enumerate(soup.find_all("div", class_="contentD")):
            if i > 4:
                break
            headline = news.text[:-26]
            date = news.text[-26:]
            output.append(f"{i+1}. {headline}. \n Dated: {date}\n\n")
    else:
        print("Failed to get the page", response.status_code)
    return output
```

```python
model = pickle.load(open("News/SentimentModel/newsSentiment.pkl", "rb"))
vectorizer = pickle.load(open("News/SentimentModel/vectorizer.pkl", "rb"))
```

```python
def sentiment(text):
    textData = vectorizer.transform([text])
    if model.predict(textData) == 0:
        print("\tNegative ðŸ“‰")
    else:
        print("\tPositive ðŸ“ˆ")
```

```python
def fetch_analyse_sentiment(name):
    news = latest_5_news(name)
    for headline in news:
        print(headline)
        sentiment(headline)
        print("\n\n")
```

```python
fetch_analyse_sentiment("google")
```
